```python
import pandas as pd
```

```python
df = pd.read_excel('/Users/bianca/Documents/GitHub/DAT-10-14/class material/Unit2/data/superstore.xls')
```

```python
grouper = df.groupby(pd.Grouper(key='Order Date', freq='M'))['Sales']
```

```python
#sum of all sales by month
grouper.sum()

#so this is the sum of the month and the 'max' of the month
grouper.sum()/grouper.max()
```

```python
#how to divide by the max sales from all the months
grouper.sum()/grouper.sum().max()
```

```python
#value_counts = how frequently something shows up
#so this is unique orders by category

df.groupby('Category')['Product ID'].value_counts()
```
