```python
sales = pd.read_excel("/Users/bianca/Documents/GitHub/DAT-10-14/class material/Unit2/data/superstore.xls")
```

```python
sales.head()
```

```python
sales = sales.set_index('Order Date')
```

```python
sales.head()

#if dealing with time series, very common to use date as the index
```

```python
sales.resample('M')['Sales'].sum()
#returns all the sales aggregated by month
#any unit of time that exists with a time stamp can be used

sales.resample('D')['Sales'].sum()
```

```python
sales.resample('min')['Sales'].sum()

#can even aggregate by the minute IF you have minute-level data (which this sales dataset does not)
```

```python
sales.groupby('Category').resample('M')['Sales'].sum()
```

```python
sales.head()

#every single transaction has a corresponding category

sales.groupby('Category')['Sales'].mean()
```

```python
# now can compare to average category sales
grouped = sales.groupby('Category')['Sales'].mean().to_frame()

sales.merge(grouped, on='Category', how='left')
```

```python
#add rename method in order to rename the new column header (vs. what will be shown which is Sales_y)
grouped = sales.groupby('Category')['Sales'].mean().to_frame().rename({'Sales': 'Cat Average'}, axis=1)

sales.merge(grouped, on='Category', how='left')
```

```python
three_mo_offset = pd.DateOffset(months=3)
```

```python
#every customer grabbing the Ship Date column and add three months to that
sales.groupby('Customer Name')['Ship Date'].min() + three_mo_offset
```
