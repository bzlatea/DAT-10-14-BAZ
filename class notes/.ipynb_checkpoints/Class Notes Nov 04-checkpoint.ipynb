```python
simple_array=np.array([1,2,3,4])
```

```python
#running total of all values
np.cumsum(simple_array)
```

```python
#running product of all values
np.cumprod(simple_array)
```

```python
(raw/raw.shift(1)).cumprod()

#percent daily fluctuations and then effectively rates of change
#multiple rates of change together, get running total
#all the way at the bottom, starting from the baseline of 1, the cummulative value was 2.34x or 2.28 times
#
```

```python
#take log of value to compress
%matplotlib inline
(raw/raw.shift(1)).cumprod().plot(logy=True)

#10 to the 0 is 1 when interpreting y-axis
```

```python
#another way
#automatically ends on Sunday

raw.resample('W').last().tail()
```

```python
#this will shift things to Friday
raw.resample('W-Fri').last().tail()
```

```python
#now see percent change from what it was last week to this week
(raw.resample('W-Fri').last()/raw.resample('W-Fri').last().shift(1))
```

```python
#so now, which had the highest average change?
(((raw.resample('W-Fri').last()/raw.resample('W-Fri').last().shift(1))-1)*100).mean(axis=1)
```

```python
#then top 3
(((raw.resample('W-Fri').last()/raw.resample('W-Fri').last().shift(1))-1)*100).mean(axis=1).nlargest(3)
```
